```python
import pandas as pd
import numpy as np
df = pd.read_csv('datasets/census.csv')
```

```python
mask_state = df['SUMLEV'] == 40
mask_county = df['SUMLEV'] == 50
```

```python
df_state = df[mask_state]
df_county = df[mask_county]
```

```python
df_state_total = df_state[['STNAME','CENSUS2010POP']].set_index('STNAME')
df_state_total.head()
```

```python
# COMPLETE THE FOLLOWING LINE
df_county_sum = df_county.groupby('STNAME').agg({'CENSUS2010POP': np.nansum})
df_county_sum = df_county.groupby('STNAME').agg(np.nansum) 
df_county_sum.head()
```

```python
df_county_sum.rename(columns = {'CENSUS2010POP':'CTY sum pop'}, inplace = True)
df_state_total.rename(columns = {'CENSUS2010POP':'ST total pop'}, inplace = True)

df_check = df_county_sum.merge(df_state_total, right_index = True, left_index = True)
df_check
```

```python
df_check['error'] = np.abs( df_check['CTY sum pop'] - df_check['ST total pop'] )
print(df_check)
```
